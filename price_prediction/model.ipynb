```python

df = pd.read_csv('train.csv')
df['size'].fillna('0 BHK',inplace = True)
#nba["College"].fillna("No College", inplace = True) 

df['size'] = list(map(lambda x: int(x.split(' ')[0]),list(df['size'])))
df.head()
```

```python
df['location'].fillna('Unknown',inplace = True)
locations = list(set(df['location']))

loc_dict = dict()

for i in locations:
    df1 = df.loc[df['location'] == i]
    prices = list(df1['price'])
    total_sqfts = [float(j) for j in list(df1['total_sqft'])]
    price_sqft_ratio = [prices[k]/total_sqfts[k] for k in range(len(prices))]
    loc_dict[i] = sum(price_sqft_ratio)/len(price_sqft_ratio)

print(loc_dict)
```
