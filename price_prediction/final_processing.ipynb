```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('t.csv')
```

```python
#area_type
area_types = list(set(df['area_type']))
area_types_dict = {}
for i in range(len(area_types)):
    area_types_dict[area_types[i]] = i+1
#area_types_dict
```

```python
final_area = list(df['area_type'])
for i in final_area:
    if i in area_types:
        k=final_area.index(i)
        final_area[k]=area_types_dict[i]
df      
```

```python
#availability
```

```python
availability = list(set(df['availability']))
availability_types = {'Immediate Possession':1,'Ready To Move':2}
```

```python
f_a = list(df['availability'])
for i in f_a:
    if i in availability:
        k=f_a.index(i)
        try:
            f_a[k]=availability_types[i]
        except:
            f_a[k]=3
```

```python

#size
df['size'].fillna('2 BHK',inplace = True)
df['size'] = list(map(lambda x: int(x.split(' ')[0]),list(df['size'])))
```

```python
#total_sqft
final_sqft = list(df['total_sqft'])
final_sqft1=[]
for i in final_sqft:
    try:
        final_sqft1.append(int(i))
    except:
        m=sum(final_sqft1)/len(final_sqft1)    
        final_sqft1.append(m)
```

```python
df['total_sqft']=final_sqft1
```

```python
#bath
final_bath = list(df['bath'])
for i in final_bath:
    k=final_bath.index(i)
    try:
        final_bath[k]=int(i)
    except:
        final_bath[k]=0
```

```python
#balcony
final_bal = list(df['balcony'])
for i in final_bal:
    k=final_bal.index(i)
    try:
        final_bal[k]=int(i)
    except:
        final_bal[k]=0
   
```

```python
#location
```

```python
df['location'].fillna('unknown',inplace = True)
```

```python
location_types = list(set(df['location']))                     
d1 = [[] for x in range(0,len(location_types))]
d2 = [[] for x in range(0,len(location_types))]
d3 = [[] for x in range(0,len(location_types))]
```

```python
dict_of = {k: v for k, v in df.groupby('location')}
dict_of
```

```python
df3=[]
for i in location_types:
    k=location_types.index(i)
    df1=list(dict_of[i]['total_sqft'])
    df2=list(dict_of[i]['price'])
    for j in range(len(df1)):
        a=int(df2[j])/int(df1[j])
        df3.append(a)    
    p=sum(df3)/len(df3)
    d3[k]=p
```

```python
dict12 = dict(zip(location_types,d3))
```

```python
import operator
sorted_d = sorted(dict12.items(), key=operator.itemgetter(1))
```

```python
p1=[]
p2=[]
for i in range(0,1306):
    p1.append(sorted_d[i][0])
    p2.append(i+1)
```

```python
final_dic = dict(zip(p1,p2))
```

```python
len(final_dic)
```

```python
final_loc = list(df['location'])
for i in final_loc:
    if i in location_types:
        k=final_loc.index(i)
        final_loc[k]=final_dic[i]
```

```python
df['area_type']=final_area
df['availability']=f_a
df['location']=final_loc
df['bath']=final_bath
df['balcony']=final_bal
```

```python
df.to_csv('final_test.csv',sep=',')
```
